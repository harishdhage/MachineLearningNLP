```python
sentence = "Hampi or Hampe (Kannada: [hɐmpe]), also referred to as the Group of Monuments at Hampi, is a UNESCO World Heritage Site located in the town of Hampi in Vijayanagara district, east-central Karnataka, India.[2] Hampi predates the Vijayanagara Empire; it is mentioned in the Ramayana and the Puranas of Hinduism as Pampa Devi Tirtha Kshetra.[3][4] Hampi continues as a religious centre, with the Virupaksha Temple, an active Adi Shankara-linked monastery and various monuments belonging to the old city."
```

```python
from nltk.tokenize import word_tokenize
import nltk
```

```python
tokenword = nltk.word_tokenize(sentence)
```

```python
tag_ele = nltk.pos_tag(tokenword)
```

```python
nltk.download('maxent_ne_chunker')
```

```python
nltk.download('words')
```

```python
nltk.ne_chunk(tag_ele)
```

```python
nltk.ne_chunk(tag_ele).draw()
```
