### Named Entity

```python
sentnc = """Money Heist (original title: "La Casa de Papel") is a Spanish television series created by Álex Pina. 
The plot revolves around a mysterious man known as the Professor, who recruits a group of eight criminals to carry out an ambitious plan to rob the Royal Mint of Spain, aiming to print billions of euros. The series is known for its intricate plot, character development, and emotional depth, making it one of the most popular non-English language series on Netflix.
Name : Harish Dhage
DOB : 22/12/1966
Place : Krakow
""".lower()
```

```python
sentnc
```

```python
import nltk
```

```python
words = nltk.word_tokenize(sentnc)
words
```

```python
postag=nltk.pos_tag(words)
```
