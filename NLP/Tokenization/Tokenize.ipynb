```python
corpus = '''Pikachu is the electric-type mouse Pokémon's species and the mascot of the Pokémon franchise. 
Created by Atsuko Nishida and finalized by Ken Sugimori, it is known for its yellow fur, red cheeks, and ability to generate electricity, which it can use in attacks or to cause thunderstorms when many gather. Pikachu evolves from Pichu when leveled up with high friendship and can evolve into Raichu when exposed to a Thunder Stone. In the Alola region, it can evolve into Alolan Raichu, which is a dual Electric/Psychic-type Pokémon.'''
```

```python
corpus
```

```python
import nltk
nltk.download('punkt_tab')
```

```python
#Paragraph -> Sentence Tokenization
#Returns a list of sentences
from nltk.tokenize import sent_tokenize
document = sent_tokenize(corpus)
print(document)
```

```python
for sentence in document:
    print(sentence)
```

```python
type(document)
```

```python
#Paragraph -> Word Tokenization
#Sentence -> Words Tokenization
#Returns a list of words
from nltk.tokenize import word_tokenize as wt
words = wt(corpus)
print(words)
```

```python
for sentence in document:
    print(wt(sentence))
```

```python
#Using WordPunct Tokenizer
#This will separate punctuation from words like commas, full stops, -, ' etc.
from nltk.tokenize import wordpunct_tokenize
```

```python
punct_words = wordpunct_tokenize(corpus)
print(punct_words)
```

```python
#Using TreebankWordTokenizer
#This tokenizer uses the conventions used in the Penn Treebank.
#It splits standard contractions, treats most punctuation as separate tokens, and handles special cases like double hyphens.
#Also dot in abbreviations are not treated as separate tokens.
from nltk.tokenize import TreebankWordTokenizer
```

```python
tokenizer = TreebankWordTokenizer()
tokenizer.tokenize(corpus)
```
