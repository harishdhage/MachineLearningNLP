### Regular Expression (RegeX)

##### It is the python technique used to find the pattern of word/symbols/digits etc.

```python
import re

# Example: Regular expression to match email addresses in a string
text = "Contact us at or support@example.com or info@company.org or for more details."
pattern = r'\b[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Z|a-z]{2,}\b'
matches = re.findall(pattern, text)
print(matches)  # Output: ['support@example.com', 'info@company.org']
```

```python
# match will find the string/part of string only from index 0
print(re.match('or',text))
```

```python
print(re.match('Contact.',text))
```

```python
# search will find the required string from any index from sentence, it will find only first occurence
print(re.search('or',text))
```

#### Regex in NLP used to cleanup the data before performing stemmining

#### 1. Custom Tokenization
NLTK's RegexpTokenizer uses regex to split text into tokens based on patterns, which can handle punctuation or special cases better than default word tokenization
This can replace or enhance nltk.word_tokenize for more control over token boundaries..

```python
import nltk
from nltk.tokenize import RegexpTokenizer

# Your sample sentence
sentence = "Hampi or Hampe (Kannada: [hɐmpe]), also referred to as the Group of Monuments at Hampi, is a UNESCO World Heritage Site located in the town of Hampi in Vijayanagara district, east-central Karnataka, India.[2] Hampi predates the Vijayanagara Empire; it is mentioned in the Ramayana and the Puranas of Hinduism as Pampa Devi Tirtha Kshetra.[3][4] Hampi continues as a religious centre, with the Virupaksha Temple, an active Adi Shankara-linked monastery and various monuments belonging to the old city."

# Regex pattern: Match words, numbers, and punctuation as separate tokens
tokenizer = RegexpTokenizer(r'\w+|[^\w\s]')
tokens = tokenizer.tokenize(sentence)
print(tokens[:10])  # Output: ['Hampi', 'or', 'Hampe', '(', 'Kannada', ':', '[', 'hɐmpe', ']', ')']
```

#### 2. Text Cleaning and Preprocessing
Use regex to remove unwanted elements like brackets, citations, or non-alphabetic characters before NER.
Apply this before tokenization to improve NER accuracy by reducing noise.

```python
import re

# Remove citations like [2], [3][4]
cleaned_sentence = re.sub(r'\[\d+\]', '', sentence)
# Remove parentheses and their contents (e.g., language notes)
cleaned_sentence = re.sub(r'\([^)]*\)', '', cleaned_sentence)
print(cleaned_sentence[:100])  # Output: "Hampi or Hampe , also referred to as the Group of Monuments at Hampi, is a UNESCO World Heritage Site..."
```

#### 3. Pattern-Based Entity Extraction
Regex can identify potential entities (e.g., locations) based on patterns, which can be combined with NLTK's POS tagging and ne_chunk for hybrid NER.
This can pre-identify entities, then pass to NLTK for tagging. For more advanced NER, integrate with spaCy or regex-based libraries like regex for Unicode support.

```python
import re

# Regex to find capitalized words (potential proper nouns like locations)
location_pattern = r'\b[A-Z][a-z]+\b'
potential_locations = re.findall(location_pattern, sentence)
print(potential_locations)  # Output: ['Hampi', 'Hampe', 'Kannada', 'UNESCO', 'World', 'Heritage', 'Site', 'Hampi', 'Vijayanagara', 'Karnataka', 'India', 'Hampi', 'Vijayanagara', 'Empire', 'Ramayana', 'Puranas', 'Hinduism', 'Pampa', 'Devi', 'Tirtha', 'Kshetra', 'Hampi', 'Virupaksha', 'Temple', 'Shankara', 'Adi']

# Filter for known location keywords (extendable)
locations = [word for word in potential_locations if word in ['Hampi', 'Karnataka', 'India', 'Vijayanagara']]
print(locations)  # Output: ['Hampi', 'Hampi', 'Vijayanagara', 'Karnataka', 'India', 'Hampi', 'Vijayanagara', 'Hampi']
```
