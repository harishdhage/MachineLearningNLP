```python
words = ["running", "runner", "ran", "easily", "fairly", "cats", "cat's", "children", "geese", "better", "best", "happily",'happiness','emotion','emotional','a large house',"The student is tired",'quickly']
```

```python
from nltk.stem import WordNetLemmatizer
import nltk
nltk.download('wordnet')
lenmatizer = WordNetLemmatizer()
```

```python
'''Pos tags:
a - adjective
r - adverb
n - noun
v - verb'''
lenmatizer.lemmatize('carefully', pos='r')
```

```python
for word in words:
    print(f"Lemmatized ({word}): {lenmatizer.lemmatize(word, pos='r')}")
```

```python
lenmatizer.morphy('running', pos='v')
lenmatizer.morphy('better', pos='a')
```
