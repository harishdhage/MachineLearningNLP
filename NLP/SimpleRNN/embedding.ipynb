# Word Embeddings with Keras Embedding Layer

## Problem Statement
Learn how to convert text sequences into dense vector representations using the Keras Embedding layer. This notebook demonstrates:
1. **One-Hot Encoding**: Convert words into one-hot vectors
2. **Padding**: Standardize sequence lengths
3. **Embedding Layer**: Transform integer sequences into dense embedding vectors
4. **Output**: Generate embedding representations for text data that can be used in neural networks

## Objective
Understand how word embeddings work as an alternative to sparse one-hot representations for NLP tasks.

```python
from tensorflow.keras.preprocessing.text import one_hot
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
```

```python
sentences = ['the cat sat on the mat', 
             'the dog sat on the log', 
             'the cat sat on the log',
             'understand meaning of words',
             'Moveie is a great way to understand the world']
sentences
```

```python
# Define vocabulary size
vocab_size = 10000
```

```python
## One-hot representation
onehot_repr = [one_hot(words, vocab_size) for words in sentences]
onehot_repr
```

```python
# word embedding representation
from tensorflow.keras.layers import Embedding
import numpy as np
sent_length = 10
embedded_docs = pad_sequences(onehot_repr, padding='pre', maxlen=sent_length)
embedded_docs
```

```python
# feature dimension
dim = 10
model = Sequential()
model.add(Embedding(vocab_size, dim, input_length=sent_length))
model.compile('adam', 'mse')
model.summary()
```

```python
output = model.predict(embedded_docs)
print(f"Output shape: {output.shape}")
output
```
