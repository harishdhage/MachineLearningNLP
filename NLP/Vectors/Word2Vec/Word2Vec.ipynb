```python
import gensim.downloader as api
wv = api.load('word2vec-google-news-300')
```

```python
vek_king = wv['King']
vek_king
```

```python
vek_king.shape
```

```python
wv['modi']
```

```python
#Here similarity() score higher the value more similarity to other word and lower the value means less similarity
wv.similarity("modi","hockey")
```

```python
wv.cosine_similarities(wv['apple'], wv['fruit', 'banana','phone','chemical'])
```

```python
# distance() is used to find the how close the other word is if score is lesser then it closer and if score is more its farther
wv.distance('apple','fruit')
```

```python
wv.distances('apple','fruit')
```

```python
v=wv['king']+wv['queen']
v
```

```python
wv.most_similar([v])
```
